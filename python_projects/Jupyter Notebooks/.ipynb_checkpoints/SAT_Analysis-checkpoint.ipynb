```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np
```

```python
sat_scores_total = pd.read_csv("school_scores.csv")
```

```python
sat_scores_total
```

```python
Year_all = sat_scores_total.loc[:,"Year"].tolist()
Year_desired = []
States_all = sat_scores_total.loc[:,"State.Code"].tolist()
States_reported = []
for i, x in enumerate(Year_all):
    if x == 2005:
        Year_desired.append(x)
        States_reported.append(States_all[i])
```

```python
Total_Math = []
Total_English = []
Total_SAT = []
i = 0
while len(Year_desired) > i:
    if Year_desired[i].__eq__(Year_desired[i]):
        Total_Math.append(sat_scores_total.loc[:,"Total.Math"].tolist()[i])
        Total_English.append(sat_scores_total.loc[:,"Total.Verbal"].tolist()[i])
    i += 1
i = 0
while len(Total_Math) > i:
    Total_SAT.append(Total_Math[i] + Total_English[i])
    i += 1

Total_SAT
```

```python
fig, ax = plt.subplots(2,3)
plots = []
for i, x in enumerate(Total_SAT):
    plots.append([States_reported[i], Total_SAT[i]])
i = 0
j = 0
k = 0
while i < len(States_reported):
    ax[j][k].bar(States_reported[i:i+10], Total_SAT[i:i+10])
    i += 10
    k += 1
    if k == 3:
        k = 0
        j += 1
```
