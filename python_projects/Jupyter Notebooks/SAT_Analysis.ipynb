```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np
```

```python
sat_scores_total = pd.read_csv("school_scores.csv")
```

```python
sat_scores_total
```

The following cells are used to find and graph the average SAT of the desired year (From 2005 to 2015)

```python
Input_Year = input("Please enter desired year between 2005 and 2015: ")
Input_Year = int(Input_Year)
while not (2005 <= Input_Year and Input_Year <= 2015):
    Input_Year = input("Please enter desired year between 2005 and 2015: ")
    Input_Year = int(Input_Year)
```

```python
Year_all = sat_scores_total.loc[:,"Year"].tolist()
```

```python
Total_Math = []
Total_English = []
Total_SAT = []
States_reported = []
i = 0
while len(Year_all) > i:
    if Input_Year.__eq__(Year_all[i]):
        Math = (sat_scores_total.loc[:,"Total.Math"].tolist()[i])
        English = (sat_scores_total.loc[:,"Total.Verbal"].tolist()[i])
        Total_SAT.append(Math + English)
        States_reported.append(sat_scores_total.loc[:,"State.Code"].tolist()[i])
    i += 1
```

```python
fig, ax = plt.subplots(4,2)
plots = []
for i, x in enumerate(Total_SAT):
    plots.append([States_reported[i], Total_SAT[i]])
i = 0
j = 0
k = 0
while i < len(States_reported):
    ax[j][k].bar(States_reported[i:i+7], Total_SAT[i:i+7])
    i += 7
    j += 1
    if j == 4:
        j = 0
        k += 1
```

The following allows the user to input the desired state by their code and find the average SAT over all the reported years

```python
Input_State = input("Enter state by code: ").upper()
while Input_State not in States_all:
    Input_State = input("Enter state by code: ").upper()
```

```python
Average_SAT = 0
count = 0
Full_State = ""
for i, x in enumerate(States_all):
    if x.__eq__(Input_State):
        Average_SAT += (sat_scores_total.loc[:,"Total.Math"].tolist()[i]) + (sat_scores_total.loc[:,"Total.Verbal"].tolist()[i])
        Full_State = sat_scores_total.loc[:,"State.Name"].tolist()[i]
        count += 1
Average_SAT /= count
```

```python
print("The average SAT score of {} is {} over the years 2005 and 2015".format(Full_State,Average_SAT))
```
